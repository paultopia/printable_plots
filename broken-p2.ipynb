```python
import plottyprint
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning) # there's a silly pandas warning to suppress
import pandas as pd
import numpy as np
from datetime import datetime
%matplotlib inline
```

```python
roldf = df = pd.read_csv("rol-scores.csv")
tsdf = pd.read_csv("unemployment.csv")
# for time series we need a column of datetime objects
tsdf["Year"] = tsdf["Year"].apply(lambda x: datetime.strptime(x, '%Y %b').date())
```

```python
box = plottyprint.boxplot([roldf["per_auto"], df["assoc_org"]],
                           ["individual freedom", "associational freedom"], 
                          "Freedoms", 
                          notch=False)
```

```python
scatter = plottyprint.scatterplot(roldf["per_auto"], 
                                  roldf["RoLScore"], 
                                  "Rule of Law vs Autonomy", 
                                  "Autonomy", 
                                  "Rule of Law")
```

```python
hist = plottyprint.histogram(roldf["RoLScore"], 
                             title="Histogram of Rule of Law Scores", 
                             add_kde=False)
```

```python
ts = plottyprint.timeseries(tsdf["Year"], 
                [tsdf["unemp"], 
                 tsdf["quit"]], 
                ["unemployment", "service work quits"], "Unemployment and Quits")
```
